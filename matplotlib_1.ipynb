```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25]) # 선 그래프 그리기기
plt.show()
```

```python
plt.scatter([1, 2, 3, 4, 5], [1, 4, 9, 16, 25]) # 산점도 그리기기
plt.show()
```

```python
x = np.random.randn(1000) # randn은 표준 정규 분포를 따르는 난수 1000개를 만드는 함수
y = np.random.randn(1000)
plt.scatter(x, y)
plt.show()
```
